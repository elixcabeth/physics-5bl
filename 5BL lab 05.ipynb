```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as fit 
```

```python
# Setup A Normal Force
N = np.array([0.42, 0.52, 0.72, 0.91, 1.11, 1.30])
N_err = np.array([0.01] * 6)

# Elizabeth Setup A Data
Fs = np.array([0.11, 0.13, 0.22, 0.30, 0.34, 0.41])
Fs_err = np.array([0.01] * 6)
```

```python
def model(x, m, c):
    return m * x + c

params, cov = fit.curve_fit(model, N, Fs, sigma=Fs_err, absolute_sigma=True)
m, c = params
m_err, c_err = np.sqrt(np.diag(cov))
print(f"Elizabeth A: μs = {m:.3f} ± {m_err:.2f}")

x_fit = np.linspace(-c / m, max(N) + 0.1, 50)
y_fit = model(x_fit, m, c)
```

```python
plt.figure()
plt.errorbar(N, Fs, yerr=Fs_err, fmt="o", capsize=5)
plt.plot(x_fit, y_fit, label=f"y = {m:.2f}x + {c:.2f}")
plt.title("Linear Regression for Setup A, Elizabeth")
plt.xlabel("Normal Force (N)")
plt.ylabel("Max Force (N)")
plt.xlim(0, 1.5)
plt.ylim(0, 0.5)
plt.grid()
plt.legend()
plt.show()
```

```python
residuals = Fs - model(N, m, c)
 
plt.figure()
plt.errorbar(N, residuals, yerr=Fs_err, fmt="o", capsize=5)
plt.title("Residuals for Setup A, Elizabeth")
plt.xlabel("Normal Force")
plt.ylabel("Residuals")
plt.ylim(-0.1, 0.1)
plt.grid()
plt.show()
```

```python
chi2 = np.sum((residuals / Fs_err) ** 2)
nu = len(N) - len(params)
chi2_red = chi2 / nu

print("Elizabeth Setup A")
print(f"Chi-squared: {chi2:.2f}")
print(f"Degrees of freedom (ν): {nu}")
print(f"Reduced Chi-squared: {chi2_red:.2f}")
```
